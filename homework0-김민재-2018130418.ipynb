```python
def calculator(a, b, operator):
    if operator == '+':
        return a + b
    elif operator == '-':
        return a - b
    elif operator == '*':
        return a * b
    elif operator == '/':
        return a / b
    elif operator == '%':
        return a % b
    else:
        return "Invalid operator"
def run():
    while True:
        print("============메뉴==============")
        print("    1. 계산기 2. 종료")
        answer = int(input("입력: "))
        if answer == 1:
            a = int(input("첫번째 수를 입력해주세요: "))
            operator = input("연산자를 입력해주세요(+,-,*,/,%): ")
            b = int(input("두번째 수를 입력해주세요: "))
            result = calculator(a, b, operator)
            print(result)
        else : 
            break
run()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
arr = np.random.randint(10, size=(10, 10)) #10 10 사이즈의 array 생성
print(arr)
count = np.bincount(arr.flatten()) #10 10 사이즈 array의 정수 빈도수를 count함 flatten함수로 한줄로 정리함
seri = pd.Series(count, name = "정수 등장 횟수") #빈도수를 series형식으로 만듬
print(seri)
plt.hist(arr.flatten(), bins=range(11)) #10개의 구간으로 나눈 그래프를 그림
plt.xlabel("number")
plt.ylabel("frequency")
plt.show()
```

```python
import numpy as np
arr = np.random.randint(0, 1000, 1000)
print(arr)
n = len(arr)
for i in range(n):
    for j in range(0, n-i-1):
        if arr[j] > arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]
print(arr)
```
